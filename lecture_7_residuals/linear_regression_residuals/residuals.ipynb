# Analyzing linear regression with residuals

Gauss theorem tells us that least squares is the "best" regression when a few hypothesis are verified. One of them is that the nois must not be biased.
Here we study a homoscedastic noise and we look at the residual.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

Below bloc generate homoscedastic data (it follows a linear function plus a normal noise which stdev is parametered by X)

```python
def generate_data(size, bounds = (0, 1000)):
    X = np.linspace(bounds[0], bounds[1], size)
    sd_deviation = np.random.uniform(0, 4, size=size) + (X / 13)
    error = np.random.normal(0, sd_deviation, size=size)
    y_train = error + X
    return y_train
```

We plot our generated data.

```python
a = generate_data(100)
x = np.linspace(0,1000,100)
print(len(a), len(x))
plt.plot(x,a, 'o')
```

```python
regress = scipy.stats.linregress(x, a)
```

```python
y = x*regress.slope + regress.intercept
plt.plot(x,a,'o')
plt.plot(x,y)
```

## Plotting the residual

```python
plt.plot(x, a-y, 'o')
```
